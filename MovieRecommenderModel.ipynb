```python
#collecting only Director  from crew
```

```python
def fetchdir(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if (i['job'] == 'Director'):
            L.append(i['name'])
            counter+=1
            break
    return L
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
# removing spaces in between words
```

```python
#movies['genres'] = movies['genres'].apply(lambda x:x.replace(" ",""))
movies['overview'] = movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# combining 5 features into one
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(10)
```

```python
movies_new = movies[['id','title','tags']]
```

```python
movies_new.head(10)
```

```python
#converting list to string again
```

```python
movies_new['tags'] = movies_new['tags'].apply(lambda x: " ".join(x) if isinstance(x, list) else str(x))
```

```python
movies_new.head(10)
```

```python
movies_new['tags'][0]
```

```python
#converting to lower cases for synschronisation
```

```python
movies_new['tags'] = movies_new['tags'].apply(lambda x:x.lower())
```

```python
movies_new.head(10)
```

```python
#converting words to vector
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(movies_new['tags']).toarray()
```

```python
#converting similar versions of words to common word e.g. dancing and danced to danc
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('dancing')
```

```python
movies_new['tags'] = movies_new['tags'].apply(stem)
```

```python
#calculate cosine distance i.e. angle between two vectors since too many dimensions
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
#keeping sorting of related movies and sorting based on score
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
#taking only top 5 movies with highest cosine similarity or lowest cosine distance
```

```python
def recommend(movie):
    movie_index = movies_new[movies_new['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies_new.iloc[i[0]].title)
```

```python
recommend('Jurassic Park')
```
