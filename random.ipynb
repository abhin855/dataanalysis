```python
import numpy as np
dataset = [10,12,15,16,103,20,18,19,22,1789,13,14,16,18,19]
outlier=[]
def detect_outlier(data):
  threshold=3
  mean=np.mean(data)
  std=np.std(data)
  for i in data:
     z_score=(i-mean)/std
     if np.abs(z_score)>threshold:
        outlier.append(i)
  return outlier  
```

```python
detect_outlier(dataset)
```
