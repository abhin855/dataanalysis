```python
#what age group is the best in the 50mi race, 30+ races
```

```python
df2.query('race_length == "50mi"').groupby('athlete_age')['average_athlete_speed'].agg(['mean','count']).sort_values('mean',ascending = False).query('count>500')
```

```python
#slower in summer or winter?
```

```python
df2['race_month']=df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df2.head(10)
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>1 else 'Winter')
```

```python
df2.groupby('race_season')['average_athlete_speed'].agg(['mean','count']).sort_values('mean',ascending = True)
```
